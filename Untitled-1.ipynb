```python

# Cargar señal ECG completa
record_path = '100'
n_samples=900
record = wfdb.rdrecord(record_path)
ecg_signal = record.p_signal[0:n_samples, 1]
fs = record.fs
t = np.arange(0, ecg_signal.shape[0]/fs, 1/fs)

# Extraer segmento con segundo ciclo QRS
index_start = np.argmin(np.abs(t - 0.985))
index_end = np.argmin(np.abs(t - 1.055))
qrs_cycle = ecg_signal[index_start:index_end]

# Calcular correlación cruzada
corr = np.convolve(ecg_signal, qrs_cycle[::-1], mode='full')
corr = corr / (np.linalg.norm(ecg_signal) * np.linalg.norm(qrs_cycle))

# Encontrar desplazamiento con máximo valor de correlación
lag = np.argmax(corr) - len(qrs_cycle) + 1

# Graficar resultados
fig, ax = plt.subplots(constrained_layout=True, figsize=(10, 4))
ax.plot(np.arange(len(ecg_signal))/fs, ecg_signal)
ax.set_title('ECG signal')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Amplitude')
ax.grid()

fig, ax = plt.subplots(constrained_layout=True, figsize=(10, 4))
ax.plot(np.arange(len(qrs_cycle))/fs, qrs_cycle)
ax.set_title('Second QRS cycle')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Amplitude')
ax.grid()

fig, ax = plt.subplots(constrained_layout=True, figsize=(10, 4))
ax.plot(np.arange(len(corr))/fs - len(qrs_cycle)/fs, corr)
ax.set_title('Cross-correlation of second cycle with ECG signal')
ax.set_xlabel('Lag (s)')
ax.set_ylabel('Correlation coefficient')
ax.grid()

print(f'Maximum correlation at lag {lag/fs:.3f} seconds.')
```

```python
import wfdb
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal

# Define the path to the .dat file
record_path = '100'

# Read the .dat file using the wfdb.rdrecord() function
record = wfdb.rdrecord(record_path)

# Access the ECG signal and other metadata from the record object
n_samples = 900
ecg_signal = record.p_signal
ecg_signal = ecg_signal[0:n_samples, 1]
fs = record.fs
t = np.arange(0, ecg_signal.shape[0]/fs, 1/fs)

# Extract segment between 0.985 and 1.055 seconds
index_start = np.argmin(np.abs(t - 0.985))
index_end = np.argmin(np.abs(t - 1.055))
segment = ecg_signal[index_start:index_end]

# Create figure and plot segment of ECG signal
fig, ax = plt.subplots(constrained_layout=True, figsize=(10, 4))
ax.plot(np.arange(len(segment))/fs, segment)
ax.set_title('ECG signal between 0.985 and 1.055 seconds')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Amplitude')
ax.grid()

# Calculate the STFT of the segment
f, tp, Zxx_kernel = signal.stft(segment, fs, nperseg=16, noverlap=8)

# Compute the absolute value of the STFT and normalize the values
abs_Zxx_kernel = np.abs(Zxx_kernel)
abs_Zxx_norm_kernel = abs_Zxx_kernel / np.max(abs_Zxx_kernel)

# Convert the normalized values to a numpy array
kernel = np.array(abs_Zxx_norm_kernel)

# Calculate the STFT of the ecg_signal
f, tp, Zxx_ecg = signal.stft(ecg_signal, fs, nperseg=16, noverlap=8)

# Compute the absolute value of the STFT and normalize the values
abs_Zxx_ecg = np.abs(Zxx_ecg)
abs_Zxx_norm_ecg = abs_Zxx_ecg / np.max(abs_Zxx_ecg)

# Convolve the ecg_signal spectrogram with the kernel using the convolve2d function
convolved_signal = signal.convolve2d(abs_Zxx_norm_ecg, kernel, mode='same')

# Create a new figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, constrained_layout=True, figsize=(10, 8))

# Plot the original ECG spectrogram in the top subplot
S = 20*np.log10(abs_Zxx_ecg)
im = ax1.pcolormesh(tp, f, S, vmin=0, vmax=np.max(S), shading='gouraud')
fig.colorbar(im, ax=ax1)
ax1.set_title('ECG spectrogram')
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Frequency (Hz)')

f, tp, Zxx = signal.stft(segment, fs, nperseg = 16, noverlap = 8)
fig, ax = plt.subplots(figsize = (15,5))
S = 20*np.log10(np.abs(segment))
im = ax.pcolormesh(tp, f, np.abs(segment), vmin=0, vmax=np.max(np.abs(segment)), shading='gouraud')
fig.colorbar(im, ax=ax)
ax.set_title('Short Time Fourier Transform')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Frequency (Hz)')

# Plot the convolved spectrogram in the bottom subplot
convolved_segment = convolved_signal[np.argmin(np.abs(f-30)):np.argmin(np.abs(f-70)), :]
S_convolved = 20*np.log10(convolved_segment)
im = ax2.pcolormesh(tp, f[np.argmin(np.abs(f-30)):np.argmin(np.abs(f-70))], S_convolved, vmin=0, vmax=np.max(S_convolved), shading='gouraud')
fig.colorbar(im, ax=ax2)
ax2.set_title('Convolved ECG spectrogram')
ax2.set_xlabel('Time (s)')

fig, ax = plt.subplots(constrained_layout = True, figsize = (10,4))
ax.plot(t,ecg_signal)
ax.set_title('Discrete time ECG signal')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Amplitude')
ax.grid()
```
